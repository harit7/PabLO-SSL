```python
import pickle

import numpy as np
```

```python
with open("args.pkl", "rb") as f:
    args = pickle.load(f)
lst_classes, inf_out, auto_lbl_conf, val_ds, val_idcs, logger, err_threshold = args
```

```python
y_val = val_ds.Y.cpu().numpy()
y_hat = inf_out["labels"].cpu().numpy()
N_v = len(y_hat)
score_type = auto_lbl_conf["score_type"]
scores = inf_out[score_type]

y_val.shape, y_hat.shape, scores.shape, type(y_val), type(y_hat), type(scores)
```

```python
list_of_thresholds = scores
```

```python
sorted_thresholds = list_of_thresholds
where_matrix = scores[:, None] >= sorted_thresholds
N_t = where_matrix.sum(axis=0)
N_e = np.sum((y_val[:, None] != y_hat[:, None]) & where_matrix, axis=0)
err_th = N_e / N_t
err_th += 0.25 * np.sqrt(1 / N_t)
good_th = list_of_thresholds[err_th < err_threshold]
good_th
```
